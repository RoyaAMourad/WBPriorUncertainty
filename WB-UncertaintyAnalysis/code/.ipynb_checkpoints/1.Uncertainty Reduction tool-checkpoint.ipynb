# **Prior Uncertainty Reduction tool**


This notebook contains code to reduce prior uncertainties using the seasonal water balances. 
An ensemble of long-term average *`ε≈ Qin-Qout`* is solved for unique *P*, *ET*, and ∆*TWS* combinations, and  disaggregated per season:

1. [monsoon season](#rainy)
1. [irrigation season](#irrigate)


<div><img src="wb_seasonal.png" width="800"></div>

* import necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import itertools
```

* Prepare data
1. *Load the CSV files into data frames*

```python
p= pd.read_csv(r"../data/CSVs/p.csv")
et= pd.read_csv(r"../data/CSVs/et.csv")
tws = pd.read_csv(r"../data/CSVs/tws.csv")
```

```python
pcoldict = {'date':'date', 
            'P_MSWEP':'MSWEP', 
            'P_GPM':'GPM',  
            'P_CHIRPS':'CHIRPS',
            'P_TRMM': "TRMM"}
pdf = p.loc[:, p.columns.isin(pcoldict.keys())]
pdf.dropna(axis = 0, how = 'all', inplace = True)
pdf.rename(columns = pcoldict, inplace = True)
pdf
```

```python
etcoldict = {'date':'date',
             'ET_SEBAL': "SEBAL",
             'ET_SSEBop':'SSEBop(a)',
             'ET_WAPOR':'WaPOR', 
             'ET_MODIS':'MOD16',  
             'ET_SSEBop_Lx':'SSEBop(b)',
             'ET_ALEXI': 'ALEXI',
             'CMRSET_ET':'CMRSET'}    
etdf= et.loc[:, et.columns.isin(etcoldict.keys())]
etdf.rename(columns = etcoldict, inplace = True)
etdf
```

```python
gcoldict = {'date':'date',
            'mascon_cri': "MAS",
            'land_jpl':'JPL', 
            'land_gfz':'GFZ',  
            'land_csr':'CSR'}    
gdf = tws.loc[:, tws.columns.isin(gcoldict.keys())]
gdf.rename(columns = gcoldict, inplace = True)
gdf
```

2. merge the dfs to master df

```python
wb_df = pdf.merge(gdf, on='date').merge(etdf, on='date')
wb_df
```

3. filter master df based on common dates

```python
wb_df['date'] = pd.to_datetime(wb_df['date'])
wb_df = wb_df[(wb_df['date'].dt.year >= 2003) & (wb_df['date'].dt.year <= 2012)]
wb_df.reset_index(drop=True, inplace=True)
print(wb_df)
```

4. create ensemble list

```python
plist = [ "MSWEP", "GPM", 'CHIRPS',"TRMM"]
etlist = [ 'SEBAL', "SSEBop(a)", 'WaPOR','MOD16','ALEXI','CMRSET',"SSEBop(b)"]
glist = ['MAS', "JPL",'GFZ','CSR']

combo_list = list(itertools.product(*[plist,etlist,glist]))
print("ensemble has {} members".format(str(len(combo_list))))
```

# 1. monsoon season <a class="anchor" id="rainy"></a> 
[Prior Uncertainty Reduction tool](#0.1)

```python
results_ro = []  
date_p = pd.to_datetime(pdf['date']).tolist()  
date_et = pd.to_datetime(etdf['date']).tolist()  
date_g = pd.to_datetime(gdf['date']).tolist()  

common_dates = wb_df['date'].tolist()

for p, et, g in combo_list:
    RO=(wb_df[g] - wb_df[p]+ wb_df[et]) 
    runoff_combos = {'date': common_dates, 'p_et_g': f"{p}_{et}_{g}", 'result': RO.tolist(), 'p':wb_df[p].tolist(),'et':wb_df[et].tolist(),'deltaS':wb_df[g].tolist()}
    results_ro.append(runoff_combos)

df = pd.DataFrame(results_ro)

new_df_RO = pd.DataFrame({
    'date': df['date'].explode(),
    'p_et_g': df['p_et_g'].explode(),
    'result': df['result'].explode(),
    'p':df['p'].explode(),
    'et':df['et'].explode(),
    'deltaS':df['deltaS'].explode()})
new_df_RO.reset_index(drop=True, inplace=True)
new_df_RO['date'] = pd.to_datetime(new_df_RO['date'])
new_df_RO['year'] = new_df_RO['date'].dt.year
new_df_RO['month'] = new_df_RO['date'].dt.month


# filter from June to September
filtered_df_RO = new_df_RO[(new_df_RO['month'] >= 6) & (new_df_RO['month'] <= 9)]
filtered_df_RO.drop(columns=['year', 'month'], inplace=True)
filtered_df_RO.reset_index(drop=True, inplace=True)
filtered_df_RO.set_index('date', inplace=True)

# group by 'p_et_g' & aggregate to annual & calculate the sum
grouped = filtered_df_RO.groupby('p_et_g')['result','p','et','deltaS'].resample('Y').sum().reset_index()
# rename
grouped.columns = ['p_et_g', 'date','annual_runoff_sum','p','et','deltaS'] 

avg_annual_runoff = grouped.groupby('p_et_g')['annual_runoff_sum','p','et','deltaS'].mean().reset_index()
avg_annual_runoff.columns = ['p_et_g', 'avg_epsilon_mm','p','et','deltaS']

# print epsilon for rainy season
avg_annual_runoff
```

# 2. irrigation season <a class="anchor" id="irrigate"></a> 
[Prior Uncertainty Reduction tool](#0.1)

```python
results_inflows = []  
date_p = pd.to_datetime(pdf['date']).tolist()  
date_et = pd.to_datetime(etdf['date']).tolist()  
date_g = pd.to_datetime(gdf['date']).tolist()  

common_dates = wb_df['date'].tolist()

for p, et, g in combo_list:
    inflows=wb_df[g] - wb_df[p]+ wb_df[et]
    inflows_combos = {'date': common_dates, 'p_et_g': f"{p}_{et}_{g}", 'result': inflows.tolist(), 'p':wb_df[p].tolist(),'et':wb_df[et].tolist(),'deltaS':wb_df[g].tolist()}
    results_inflows.append(inflows_combos)

df_inflows = pd.DataFrame(results_inflows)

new_df_inflows = pd.DataFrame({
    'date': df_inflows['date'].explode(),
    'p_et_g': df_inflows['p_et_g'].explode(),
    'result': df_inflows['result'].explode(),
    'p':df_inflows['p'].explode(),
    'et':df_inflows['et'].explode(),
    'deltaS':df_inflows['deltaS'].explode()})
new_df_inflows.reset_index(drop=True, inplace=True)
new_df_inflows['date'] = pd.to_datetime(new_df_inflows['date'])
new_df_inflows['year'] = new_df_inflows['date'].dt.year
new_df_inflows['month'] = new_df_inflows['date'].dt.month

#filter df for months outside monsoon
rabi_df = new_df_inflows[((new_df_inflows['month'] >= 1) & (new_df_inflows['month'] <= 5)) | ((new_df_inflows['month'] >= 10) & (new_df_inflows['month'] <= 12))]
rabi_df.drop(columns=['year', 'month'], inplace=True)
rabi_df.reset_index(drop=True, inplace=True)
rabi_df['date'] = pd.to_datetime(rabi_df['date'])
rabi_df.set_index('date', inplace=True)

# group by 'p_et_g' &aggregate to annual 
rabi_df_grouped = rabi_df.groupby('p_et_g')['result','p','et','deltaS'].resample('Y').sum().reset_index()
# rename
rabi_df_grouped.columns = ['p_et_g', 'date','annual_inflows_sum','p','et','deltaS']


avg_annual_inflow = rabi_df_grouped.groupby('p_et_g')['annual_inflows_sum','p','et','deltaS'].mean().reset_index()
avg_annual_inflow.columns = ['p_et_g', 'avg_epsilon_mm','p','et','deltaS']
# print epsilon for irrigation season
avg_annual_inflow
```
